```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
Consumo = np.array([50,140,30,75,123]).reshape(-1,1)
Demanda = np.array([100,200,80,100,160])

modelo = LinearRegression()
modelo.fit(Consumo,Demanda)
```

```python
#A = np.array([15, 25, 33]).reshape(-1, 1) # vazão de água em m3/min
ConsumoTESTE = np.array([35,94,28,47,111]).reshape(-1, 1) # vazão de água em m3/min
# Fazer previsões com base na vazão de água
Demanda_previsto = modelo.predict(ConsumoTESTE)

#imprimir as revisoes
for i in range(len(ConsumoTESTE)):
 print(f"Consumo: {ConsumoTESTE[i][0]} Demanda: {Demanda_previsto[i]}")
plt.scatter(Consumo,Demanda
            , color='blue', label='Dados de Treinamento')
plt.plot(Consumo,Demanda_previsto, color = 'red',label='Dados de teste')
plt.xlabel('Consumo')
plt.ylabel('Demanda')
plt.title('Relação entre consumo e demanda')
plt.legend()
plt.grid(True)
plt.show()

```
