```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
A = np.array([50,100,120,300]).reshape(-1,1)
VALOR = np.array([180.000,300.000,375.000,600.000])

modelo = LinearRegression()
modelo.fit(A,VALOR)
```

```python
#A = np.array([15, 25, 33]).reshape(-1, 1) # vazão de água em m3/min
ATESTE = np.array([25,125,452,200]).reshape(-1, 1) # vazão de água em m3/min
# Fazer previsões com base na vazão de água
VALOR_previsto = modelo.predict(ATESTE)

#imprimir as revisoes
for i in range(len(ATESTE)):
 print(f"Área: {ATESTE[i][0]} Valor: {VALOR_previsto[i]}")
plt.scatter(A,VALOR, color='blue', label='Dados de Treinamento')
plt.plot(A,VALOR_previsto, color = 'red',label='Dados de teste')
plt.xlabel('Área')
plt.ylabel('Valor')
plt.title('Relação entre área e valor')
plt.legend()
plt.grid(True)
plt.show()

```
