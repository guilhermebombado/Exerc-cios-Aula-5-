```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
Renda = np.array([1200,5000,8000,10000,30000]).reshape(-1,1)
Gastos = np.array([750,1500,4000,6000,18000])

modelo = LinearRegression()
modelo.fit(Renda,Gastos)
```

```python
#A = np.array([15, 25, 33]).reshape(-1, 1) # vazão de água em m3/min
RendaTESTE = np.array([800,2500,7000,5890,30000]).reshape(-1, 1) # vazão de água em m3/min
# Fazer previsões com base na vazão de água
Gastos_previsto = modelo.predict(RendaTESTE)

#imprimir as revisoes
for i in range(len(RendaTESTE)):
 print(f"Renda: {RendaTESTE[i][0]} Gastos: {Gastos_previsto[i]}")
plt.scatter(Renda,Gastos
            , color='blue', label='Dados de Treinamento')
plt.plot(Renda,Gastos_previsto, color = 'red',label='Dados de teste')
plt.xlabel('Renda')
plt.ylabel('Gastos')
plt.title('Relação entre renda e gastos')
plt.legend()
plt.grid(True)
plt.show()

```
