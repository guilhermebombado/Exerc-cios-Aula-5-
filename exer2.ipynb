```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
Q = np.array([100,200,300,400,500,600,700]).reshape(-1,1)
T = np.array([30,35,40,45,60,100,150])

modelo = LinearRegression()
modelo.fit(Q,T)
```

```python
#Q = np.array([15, 25, 33]).reshape(-1, 1) # vazão de água em m3/min
QTESTE = np.array([108,405,360,520,685,850,524]).reshape(-1, 1) # vazão de água em m3/min
# Fazer previsões com base na vazão de água
T_previsto = modelo.predict(QTESTE)

#imprimir as revisoes
for i in range(len(QTESTE)):
 print(f"Quantidade de calor: {QTESTE[i][0]} cal temp: {T_previsto[i]} metros")
plt.scatter(Q,T, color='blue', label='Dados de Treinamento')
plt.plot(Q,T_previsto, color = 'red',label='Dados de teste')
plt.xlabel('Quantidade de calor')
plt.ylabel('Temperatura')
plt.title('Relação entre quantidade de calor e temperatura')
plt.legend()
plt.grid(True)
plt.show()

```
